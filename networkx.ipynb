```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
n = 10 # number of nodes
p = (0)/n # probability of each two nodes being connected
print(p)
G = nx.fast_gnp_random_graph(n,p)
nx.draw(G, with_labels=True, font_weight='bold')
```

```python
ps = np.linspace(0.0, 1.0, 100) 
l = np.zeros(len(ps))
Ncs = np.zeros(len(ps))
for k in ps:
    l[k] = ps[k] / n
    Ncs[k] = nx.number_connected_components(l[k])
return Ncs

#Ncs = nx.number_connected_components(G)
print(len(ps))
print(Ncs)
```
