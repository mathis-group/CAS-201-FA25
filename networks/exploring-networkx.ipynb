```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
n = 100 # Number of nodes
ps = np.linspace(0.0, 1.0, 10) # Array of size 10 uniformly spaced between 0 and 1
Ncs = np.random.random(10) # Completely random synthetic data TODO we need to replace this with real data
# plt.scatter(ps, Ncs)

def mean_degree(my_graph):
    degree_sequence = dict(my_graph.degree())
    mean_degree = np.mean([v for v in degree_sequence.values()])
    return mean_degree
```

```python
m_samples = 50
# ps = np.linspace(0.0, 0.10, 100) 
ps = np.logspace(-4, 0.0, 100)
# print(ps)

Ncs = np.zeros((len(ps), m_samples)) # The number of connected components
ave_deg = np.zeros((len(ps), m_samples)) # The average degree of nodes
# print(Ncs)
for j in range(m_samples):
    for i in range(len(ps)):
        # Generate a graph
        G = nx.fast_gnp_random_graph(n, ps[i])
        # Measure stats on it
        ## Get number of connected components
        nc = nx.number_connected_components(G)
        Ncs[i,j] = nc
        
        ## Get the mean degree of all nodes
        ave_d = mean_degree(G)
        ave_deg[i,j] = ave_d

ave_ncs = np.mean(Ncs, axis=1)
ave_ave_d = np.mean(ave_deg, axis=1)

# Ncs = Ncs / m_samples
plt.scatter(ps, ave_ncs)
plt.yscale("log")
plt.xscale("log")
# plt.hlines(y=1.0, xmin=0, xmax=1.0)
```
