```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def logistic(r, x):
    return r*x*(1.0 - x)
    
```

```python
r1 = 0.2*np.ones(25)
r2 = 1.0*np.ones(25)
# r3 = 2.5*np.ones(25)
r4 = 3.67*np.ones(25)
rs = np.concat([r1,r2,r4])
x0 = np.random.rand(len(rs))
```

```python
logistic(rs, x0)
```

```python
def run_logistic(r, x0, time_steps):
    xs = np.zeros((time_steps+1, len(x0)))
    xs[0, :] = x0
    for i in range(1,time_steps+1):
        new_xs = logistic(r, x0)
        xs[i, :] = new_xs
        x0 = new_xs
    return xs
```

```python
np_xs = run_logistic(rs, x0, 3)
```

```python
plt.plot(np_xs, marker="o") 
```

```python
# fake data
data = np.random.randn(21, 100)
r = np.linspace(0, 1, 100)  # values to map colors

x = np.arange(data.shape[0])   # 21 time points
cmap = plt.cm.viridis

# Normalize r so it maps to [0,1]
norm = plt.Normalize(r.min(), r.max())

for i in range(data.shape[1]):
    plt.plot(x, data[:, i], color=cmap(norm(r[i])))

# plt.colorbar(plt.cm.ScalarMappable(norm=norm, cmap=cmap), label="r value")
plt.show()
```
