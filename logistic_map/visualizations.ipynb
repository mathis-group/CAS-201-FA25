```python
r = 3.66995
time_steps = 100
t = np.arange(time_steps+1)
epsilion = 0.000000001

# Getting random initial conditions
x0s = np.random.random(1000)
# A small modification of those
x0s_e = x0s + epsilion

xs_original = run_logistic_arrays(r, x0s, time_steps)
xs_perturbed = run_logistic_arrays(r, x0s_e, time_steps)

delta = abs(xs_original-xs_perturbed)

# plt.plot(t, delta.transpose())
# plt.yscale("log")


```
